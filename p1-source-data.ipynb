```python
import crime as cr
import pandas as pd, numpy as np
```

```python
names = [
    'crime_16_19',
    'crime_97_15',
    'dist_arrests',
    'dist_student_mobility',
    'dist_grad_rate',
    'census_field_desc',
    'county_population',
]
ids = [cr.metadata(name)['id'] for name in names]
```

```python
for name, ID in zip(names, ids):
    df = cr.load(name, full=True)
    print(name)
    display(df.head())
    print()
    df.to_csv(f'input-data/{ID}_{name}.csv', index=False)
```
