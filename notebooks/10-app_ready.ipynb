```python
import pandas as pd, numpy as np
import df_util
from df_util import head
work_path = lambda name: f'../working-data/{name}.csv'
read = lambda name: pd.read_csv(work_path(name))
```

```python
df_county = read('everything_grouped')
df_dist = read('education_dist')
df_geo = read('geo_county_school')
df_geo_dist = df_geo[['county', 'dist', 'geo_county', 'geo_dist', 'geo_county_point', 'geo_dist_point']]
head(df_county, df_dist, df_geo_dist)
```

```python
df_dist = (df_geo_dist
    .merge(df_dist, on=['county', 'dist'])
    [ ~ df_geo_dist.geo_dist.isna()]
)

head(df_county, df_dist)
```

```python
df_county.columns = [c if '__' not in c else c.split('__')[1] for c in df_county.columns]
df_county = df_county.loc[:, ~ df_county.columns.duplicated()]
head(df_county)
```

```python
df_county.to_csv(work_path('everything_grouped_ready'), index=False)
```

```python
df_dist.to_csv(work_path('education_dist_ready'), index=False)
```
